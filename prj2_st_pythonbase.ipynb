### 캡슐화
* (내용) 객체의 세부 내용을 외부로부터 은닉하는 방법
* (방법) 변수 앞에 '__'(2개)를 붙임 --> 외부에서 접근 불가
* 은닉변수에 접근하는 방법 --> 메서드 활용
  * getter: 획득자= 은닉 값을 가져옴
  * setter: 지정자= 은닉 값을 수정
* 따라서 변수를 직접접근 없이 메서드를 통해서 수정
* 획득이 되기 때문에 변수이름을 할 수 없고, 직접 수정도 안됨
* 변수 보안이나, 객체 생성후 잘못된 접근을 통해 과정이 꼬이는 일을 방지
